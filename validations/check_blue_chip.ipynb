```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.data import get_blue_chip_returns
from src.fire import FireOptimizer

returns = get_blue_chip_returns(years=15, n_assets=10)
```

```python
# check the moments of the returns
import numpy as np
mean_returns = np.mean(returns, axis=0)
cov_matrix = np.cov(returns, rowvar=False)
print("Mean Returns:", mean_returns)
print("std:", np.sqrt(np.diag(cov_matrix)))
```
